```python
from youtube_transcript_api import YouTubeTranscriptApi
from youtube_transcript_api._errors import TranscriptsDisabled

from langchain_text_splitters import RecursiveCharacterTextSplitter

from langchain_openai import OpenAIEmbeddings, ChatOpenAI

from langchain_community.vectorstores import FAISS

from langchain_core.prompts import PromptTemplate
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
from urllib.parse import urlparse, parse_qs
from youtube_transcript_api import YouTubeTranscriptApi

url = input("Enter YouTube URL: ")

parsed = urlparse(url)

if "youtu.be" in parsed.netloc:
    video_id = parsed.path[1:]
else:
    video_id = parse_qs(parsed.query)["v"][0]

api = YouTubeTranscriptApi()

transcript = api.fetch(video_id)

text = " ".join(i.text for i in transcript)

print("\nTranscript:\n", text[:1000],"...")
```

```python
transcript[:5]
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

chunks = splitter.create_documents([text])
```

```python
len(chunks)
```

```python
chunks[50]
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vector_store = FAISS.from_documents(chunks, embeddings)
```

```python
list(vector_store.index_to_docstore_id.items())[:10]
```

## Retrieval

```python
retriever = vector_store.as_retriever(search_type="similarity", search_kwargs={"k": 4})
```

```python
retriever
```

```python
retriever.invoke('Why is scientific progress hard to recognize?')
```

## Augmentation

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()


class NvidiaLLM:
    def __init__(
        self,
        model="meta/llama-3.1-8b-instruct"
    ):
        self.client = OpenAI(
            base_url="https://integrate.api.nvidia.com/v1",
            api_key="xyz"
        )

        self.model = model

    def invoke(self, prompt):
        response = self.client.chat.completions.create(
            model=self.model,
            messages=[
                {
                    "role": "user",
                    "content": str(prompt)
                }
            ],
            temperature=0.7,
            max_tokens=1024
        )

        return response.choices[0].message.content


llm = NvidiaLLM()
```

```python
prompt = PromptTemplate(
    template="""
      You are a helpful assistant.
      Answer ONLY from the provided transcript context.
      If the context is insufficient, just say you don't know.

      {context}
      Question: {question}
    """,
    input_variables = ['context', 'question']
)
```

```python
question          = "Why is scientific progress considered difficult to recognize or verify, even in major discoveries like special relativity?"
retrieved_docs    = retriever.invoke(question)
```

```python
retrieved_docs
```

```python
context_text = "\n\n".join(doc.page_content for doc in retrieved_docs)
context_text
```

```python
final_prompt = prompt.invoke({"context": context_text, "question": question})
```

```python
final_prompt
```

## Generation

```python
answer = llm.invoke(final_prompt)
print(answer)
```

## Building a Chain

```python
from langchain_core.runnables import RunnableParallel, RunnablePassthrough, RunnableLambda
from langchain_core.output_parsers import StrOutputParser
```

```python
def format_docs(retrieved_docs):
  context_text = "\n\n".join(doc.page_content for doc in retrieved_docs)
  return context_text
```

```python
parallel_chain = RunnableParallel({
    'context': retriever | RunnableLambda(format_docs),
    'question': RunnablePassthrough()
})
```

```python
parallel_chain.invoke('Why is scientific progress hard to reduce to a fixed process?')
```

```python
parser = StrOutputParser()
```

```python
from langchain_core.runnables import RunnableLambda

main_chain = (
    parallel_chain
    | prompt
    | RunnableLambda(lambda x: llm.invoke(x.to_string()))
    | parser
)
```

```python
main_chain.invoke('Can you summarize the video')
```
